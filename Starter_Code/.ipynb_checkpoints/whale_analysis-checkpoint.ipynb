# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:
1. whale_returns.csv
2. algo_returns.csv
3. sp500_history.csv

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_returns = Path("Resources/whale_returns.csv")
whale = pd.read_csv(whale_returns, index_col='Date', infer_datetime_format=True, parse_dates=True)
whale.head()
```

```python
# Count nulls
whale.isnull().sum()
```

```python
# Drop nulls
whale.dropna(inplace=True)
whale.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns = Path("Resources/algo_returns.csv")
algo = pd.read_csv(algo_returns, index_col='Date', infer_datetime_format=True, parse_dates=True)
algo.head()
```

```python
# Count nulls
algo.isnull().sum()
```

```python
# Drop nulls
algo.dropna()
algo.isnull().sum()
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_history = Path("Resources/sp500_history.csv")
sp500 = pd.read_csv(sp500_history, index_col='Date', infer_datetime_format=True, parse_dates=True)
sp500.head()
```

```python
# Check Data Types
sp500.dtypes
```

```python
# Fix Data Types

# Convert the string into a number using replace
sp500['Close'] = sp500['Close'].str.replace('$','')

# Convert the data type to float
sp500['Close'] = sp500['Close'].astype(float)

# Check the data type
sp500.dtypes
```

```python
# Calculate Daily Returns
sp500.sort_index(inplace=True)

daily_sp500 = sp500.pct_change()
daily_sp500.head()
```

```python
# Drop nulls
daily_sp500.dropna(inplace=True)
daily_sp500.head()
```

```python
# Rename Column
daily_sp500.rename(columns={'Close':'S&P 500'}, inplace=True)
daily_sp500.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame
merge = pd.concat([whale, algo, daily_sp500], axis='columns', join='inner')
merge.head()
```

# Portfolio Analysis

In this section, you will calculate and visualize performance and risk metrics for the portfolios.

## Performance

Calculate and Plot the daily returns and cumulative returns. Does any portfolio outperform the S&P 500? 

```python
# Plot daily returns
merge.plot(figsize=(20,10), title='Daily Returns of Whale, Algorithms, and S&P 500')
```

```python
# Plot cumulative returns
cumulative = (1 + merge).cumprod() - 1
cumulative.plot(figsize=(20,10), title='Cumulative Returns of Whale, Algorithms, and S&P500')
```

## Risk

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios
4. Determine which portfolios are riskier than the S&P 500
5. Calculate the Annualized Standard Deviation

```python
# Box plot to visually show risk
merge.boxplot(grid=False, figsize=(20,10))
```

```python
# Daily Standard Deviations
std = merge.std()
std
```

```python
# Determine which portfolios are riskier than the S&P 500
std.apply(lambda portfolio: portfolio > std[6])
```

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_std = std * np.sqrt(252)
annualized_std
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Plot the rolling standard deviation of the various portfolios along with the rolling standard deviation of the S&P 500 (consider a 21 day window). Does the risk increase for each of the portfolios at the same time risk increases in the S&P?
2. Construct a correlation table for the algorithmic, whale, and S&P 500 returns. Which returns most closely mimic the S&P?
3. Choose one portfolio and plot a rolling beta between that portfolio's returns and S&P 500 returns. Does the portfolio seem sensitive to movements in the S&P 500?
4. An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the ewm with a 21 day half-life.

```python
# Calculate and plot the rolling standard deviation
# for the S&P 500 and whale portfolios using a 21 trading day window
merge.rolling(window=21).std().plot(figsize=(20,10), title='21-Day Rolling All')
```

```python
# Construct a correlation table
correlation = merge.corr()
correlation
```

```python
# Calculate Beta for a single portfolio compared to the total market (S&P 500)
# (Your graph may differ, dependent upon which portfolio you are comparing)


# Beta calculation using covariance and variance

berkshire_cov = merge['BERKSHIRE HATHAWAY INC'].cov(merge['S&P 500'])
berkshire_var = merge['S&P 500'].var()

berkshire_beta = (berkshire_cov / berkshire_var)
berkshire_beta
```

```python
# Rolling beta calculation and plot

rolling_cov = merge['BERKSHIRE HATHAWAY INC'].rolling(window=60).cov(merge['S&P 500'])
rolling_var = merge['S&P 500'].rolling(window=60).var()
rolling_beta = rolling_cov / rolling_var

rolling_beta.plot(figsize=(20,10), title='Rolling beta between Berkshire returns and S&P 500 returns')
```

```python
# Calculate a rolling window using the exponentially weighted moving average. 
merge.ewm(halflife=21).std().plot(figsize=(30,10))
```

## Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)

1. Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot.
2. Determine whether the algorithmic strategies outperform both the market (S&P 500) and the whales portfolios.

```python
# Calculate annualized Sharpe Ratios
sharpe = (merge.mean() * 252) / (merge.std() * np.sqrt(252))
sharpe
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe.plot(kind='bar', title='Sharpe Ratio')
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Read the first stock
apple = pd.read_csv('Resources/aapl.csv', index_col='Date', infer_datetime_format=True, parse_dates=True)
apple.head()
```

```python
# Read the second stock
tesla = pd.read_csv('Resources/tsla.csv', index_col='Date', infer_datetime_format=True, parse_dates=True)
tesla.head()
```

```python
# Read the third stock
google = pd.read_csv('Resources/goog.csv', index_col='Date', infer_datetime_format=True, parse_dates=True)
google.head()
```

```python
# Concatenate all stocks into a single DataFrame
portfolio = pd.concat([apple, tesla, google], axis='columns', join='inner')
portfolio.columns = ['AAPL', 'TSLA', 'GOOG']
portfolio.head()
```

```python
# Reset the index
portfolio.index = portfolio.index.normalize()
portfolio.reset_index(inplace=True)
portfolio.head()
```

```python
# Pivot so that each column of prices represents a unique symbol
portfolio = pd.pivot_table(portfolio, index=['Date'])
portfolio.head()
```

```python
# Drop Nulls & Daily Returns
portfolio_daily = portfolio.pct_change().dropna()
portfolio_daily.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns
weights = [1/3, 1/3, 1/3]

custom_returns = (portfolio_daily * weights).sum(axis=1)

custom_returns
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Add your "Custom" portfolio to the larger dataframe of fund returns
all_portfolio = pd.concat([merge, custom_returns], axis='columns', join='inner')
all_portfolio.columns = [*all_portfolio.columns[:-1], "Custom"]
all_portfolio.head()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
all_portfolio.dropna().isnull().sum()
```

## Re-run the performance and risk analysis with your portfolio to see how it compares to the others

```python
# Risk
all_std = all_portfolio.std()
all_std
```

```python
# Annualized standard deviation (252 trading days) for all portfolios
annualized_std = all_std * 252
annualized_std
```

```python
# Rolling standard deviation using a 21 trading day window for all portfolios
all_portfolio.rolling(window=21).std().plot(figsize=(20,10), title='21-Day Rolling Standard Deviation')
```

```python
# Annualized Sharpe Ratios
all_sharpe = (all_portfolio.mean() * 252) / (all_std * np.sqrt(252))
all_sharpe
```

```python
# Visualize the sharpe ratios as a bar plot
all_sharpe.plot(kind='bar', figsize=(10,5), title='Sharpe Ratios')
```

```python
# Create a correlation analysis
all_portfolio.corr()
```

```python
# Beta between all portfolio and S&P 500
all_cov = all_portfolio['Custom'].cov(all_portfolio['S&P 500'])
all_var = all_portfolio['S&P 500'].var()

all_beta = all_cov / all_var
all_beta
```

```python
# Rolling beta between all portfolio and S&P 500
rolling_all_cov = all_portfolio['Custom'].rolling(window=30).cov(all_portfolio['S&P 500'])
rolling_all_var = all_portfolio['S&P 500'].rolling(window=30).var()
rolling_all_beta = rolling_all_cov / rolling_all_var

rolling_all_beta.plot(figsize=(20,10), title='Rolling beta between all portfolio and S&P 500')
```
